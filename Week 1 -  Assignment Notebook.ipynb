# I. Data Coding and Merging

```python
# Load packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load game logs data
Teams = pd.read_excel("Game logs 1999-2003.xlsx")
```

create a dummy variable for home wins (1 or 0) and visitor wins, and create the ‘year’ variable.

```python
Teams['hwin'] = np.where(Teams['home_score'] > Teams['visitor_score'],1,0)
Teams['awin'] = np.where(Teams['home_score'] < Teams['visitor_score'],1,0)
Teams['year'] = Teams['date'].astype(str).str[0:4]
```

Create a dataframe that sums the statistics you need by season (year) and by home team. Rename the column for the home team as ‘Team’.

```python
Teamshome = Teams.groupby(['home','year'])['visitor_ab', 'visitor_h', 'visitor_2b', 'visitor_3b', 'visitor_hr', 'visitor_sf',
                                           'visitor_hbp','visitor_bb','home_ab', 'home_h', 'home_2b', 'home_3b',
                                           'home_hr', 'home_sf', 'home_hbp', 'home_bb', 'hwin','awin'].sum()
Teamshome = Teamshome.reset_index()
Teamshome = Teamshome.rename(columns = {"home": "Team"}) 
```

Create a dataframe that sums the statistics you need by season (year) and by visiting team. Rename the column for the visiting team as ‘Team’.

```python
Teamsaway = Teams.groupby(['visitor','year'])['visitor_ab', 'visitor_h', 'visitor_2b', 'visitor_3b', 'visitor_hr', 'visitor_sf',
                                           'visitor_hbp','visitor_bb','home_ab', 'home_h', 'home_2b', 'home_3b',
                                           'home_hr', 'home_sf', 'home_hbp', 'home_bb', 'hwin','awin'].sum()
Teamsaway = Teamsaway.reset_index()
Teamsaway = Teamsaway.rename(columns = {"visitor": "Team"}) 
```

Merge the two dataframes on ‘year’ and ‘Team’ and create a variable for the total number of wins in the season for each team

```python
Teams2 = pd.merge(Teamshome, Teamsaway, on =['Team','year'])
Teams2['wins'] = Teams2['hwin_x']+Teams2['awin_y']
```

Which season had the highest median number of team wins?

```python
Teams2.groupby(["year"])["wins"].median().sort_values(ascending=False).head()
```

Aggregate (sum) the number of home team hits for every individual season in the data. Which season did the minimum aggregate home team hit count occur?

```python
Teamshome.groupby(["year"])["home_h"].min().sort_values(ascending=True).head()
```

Rank the seasons from highest to lowest average number of at bats for the away teams aggregated over the entire season. 

```python
Teamsaway.groupby(["year"])["visitor_ab"].mean().sort_values(ascending=False).head()
```

# II. Calculating Performance Metrics

Create batting average for and batting average against as the sum the relevant statistics for each team as home team and visiting team.

```python
Teams2['AVGFOR'] = (Teams2['home_h_x'] + Teams2['visitor_h_y'])/(Teams2['home_ab_x'] + Teams2['visitor_ab_y'])
Teams2['AVGAGN'] = (Teams2['home_h_y'] + Teams2['visitor_h_x'])/(Teams2['home_ab_y'] + Teams2['visitor_ab_x'])
```

Create the win percentage variable for each team by generating a count of games played and dividing total wins by total games played

```python
# Calculate win percentage
Teams3 = Teams2[['year', 'Team', 'wins','AVGFOR','AVGAGN']]

TeamsGh = Teams.groupby(['year','home'])['hwin'].count()
TeamsGh = TeamsGh.reset_index()
TeamsGh = TeamsGh.rename(columns = {"home": 'Team', "hwin": 'hgame'}) 

TeamsGa = Teams.groupby(['year','visitor'])['awin'].count()
TeamsGa = TeamsGa.reset_index()
TeamsGa = TeamsGa.rename(columns = {"visitor": 'Team', "awin": 'agame'}) 

TeamsG = pd.merge(TeamsGh, TeamsGa, on =['Team','year'])
TeamsG['Games'] = TeamsG['hgame']+TeamsG['agame']
TeamsG = TeamsG[['year','Team','Games']]

Teams3 = pd.merge(Teams3, TeamsG, on =['Team','year'])
Teams3['wpc'] = Teams3['wins']/Teams3['Games']
Teams3
```

Create the win percentage variable for each team by generating a count of games played and dividing total wins by total games played.

```python
# Calculate home win percentage and away win percentage
TeamsWh = Teams.groupby(['year','home'])['hwin'].sum()
TeamsWh = TeamsWh.reset_index()
TeamsWh = TeamsWh.rename(columns = {"home": 'Team'}) 
Teamsh = pd.merge(TeamsGh, TeamsWh, on=['Team','year'])
Teamsh['hwpc'] = Teamsh['hwin']/Teamsh['hgame']

TeamsWa = Teams.groupby(['year','visitor'])['awin'].sum()
TeamsWa = TeamsWa.reset_index()
TeamsWa = TeamsWa.rename(columns = {"visitor": 'Team'})
Teamsa = pd.merge(TeamsGa, TeamsWa, on=['Team','year'])
Teamsa['awpc'] = Teamsa['awin']/Teamsa['agame']
```

calculate each team’s home and away win percentage separately for each season.Merge each team’s home and away win percentage for each season into one data frame. Calculate WPCT diff defined as home win percentage – away win percentage for every team in each season.

```python
# Merge home and away win percentage data frames and calculate difference
Teamsha = pd.merge(Teamsh, Teamsa, on=['Team','year'])
Teamsha['wpc_diff'] = Teamsha['hwpc'] - Teamsha['awpc']
Teamsha
```

```python
# Maximum WPCT diff
Teamsha.sort_values(by=['wpc_diff'], ascending = False)
```

```python
# AVG Diff
Teams3['AVGDIFF'] = Teams3['AVGFOR'] - Teams3['AVGAGN']
Teams3.sort_values(by=['AVGDIFF'])
```

# III. Running Regressions

Run an OLS (Ordinary Least Squares) regression of win percentage on batting average for and batting average against.

```python
import statsmodels.formula.api as smf
WinOBP_lm = smf.ols(formula = 'wpc ~ AVGFOR + AVGAGN', data=Teams3).fit()
WinOBP_lm.summary()
```

Run an OLS (Ordinary Least Squares) regression of win percentage on batting average for and batting average against,   constraining the coefficient on each variable to be equal and opposite.

```python
# Regression of winning percentage on batting average for and against (constrained coef)
WinAVGR_lm = smf.ols(formula = 'wpc ~ I(AVGFOR - AVGAGN)', data=Teams3).fit()
WinAVGR_lm.summary()
```
